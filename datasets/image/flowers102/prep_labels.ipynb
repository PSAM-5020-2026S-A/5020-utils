```python
import json
import tensorflow_datasets as tfds
```

```python
with open ("Oxford-102_Flower_dataset_labels.txt", "r") as ifp:
  labels = [line.strip()[1:-1].replace("'", "").replace("?", "").replace(" ", "-") for line in ifp]

len(labels)
```

```python
flowers_tfd = tfds.load("oxford_flowers102")
```

```python
file2label = {}

for split in flowers_tfd:
  for img in flowers_tfd[split]:
    file_name = img["file_name"].numpy().decode("utf-8")
    label_id = int(img["label"].numpy())
    label = labels[label_id]
    file2label[file_name] = label
    
len(file2label)
```

```python
with open("flowers102_labels.json", "w") as ofp:
  json.dump(file2label, ofp, separators=(",",":"), ensure_ascii=False)
```
